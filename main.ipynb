Assume that Joy on Earth follows distribution A with the following PDF
$$
f(x \mid \theta)=\frac{1}{\theta}, \quad 0 \leq x \leq \theta ; \quad \theta>0.
$$

(a) The maximum likelihood estimator for the parameter $\theta$, i.e., $\hat{\theta}_{\text{MLE}}$ can be obtained as follows:

The likelihood is
$$
f(\textbf{x} \mid \theta)=f(x_1,x_2,...,x_n \mid \theta)=\prod_{i=1}^n f(x_i \mid \theta) = \left(\frac{1}{\theta}\right)^n.
$$
The negative log likelihood is 
$$
L(\textbf{x} \mid \theta)=-\log(f(\textbf{x} \mid \theta)) = -\log\left(\left(\frac{1}{\theta}\right)^n\right) = n\log(\theta).
$$
When we attempt to find the stationary point of the negative log likelihood we find that the derivative with respect to $\theta$ is
$$
\frac{dL(\textbf{x} \mid \theta)}{d\theta}=\frac{dn\log(\theta)}{d\theta}=\frac{n}{\theta}.
$$
We can't set this to zero to find the stationary point and solve for the estimate of $\theta$ because there is no stationary point. Instead we have to recognise that all of our realisations $x_1,x_2,...,x_n$ will be less than or equal to $\theta$ according to the PDF definition. This is equivalent to saying $\theta \geq x_1,x_2,...,x_n$. This means the negative log likelihood is actually defined with a specific domain as follows:
The negative log likelihood is 
$$
L(\textbf{x} \mid \theta)= n\log(\theta), \quad \theta \geq x_1,x_2,...,x_n.
$$
This is an increasing function with positive slope because $n$ and $\theta$ are positive and so the derivative of the function is also positive. We can also note the domain of this function starts at $\theta$ equal to the maximum value of $x_1,x_2,...,x_n$ since $\theta \geq x_1,x_2,...,x_n$. Therefore, since this is an increasing function starting from $\theta$ equal to the maximum value of $x_1,x_2,...,x_n$, the minimum of this function occurs when $\theta$ equals the maximum value of $x_1,x_2,...,x_n$. Therefore we let the MLE estimate be $\hat{\theta}=\text{argmax}_{x_i}\{x_1,x_2,...,x_n\}$. (Draw a diagram of the negative log likelihood if you don't get it.)

(b) Now using our sample of drop bear joy $S_1$ given above we can note that the MLE estimate in this case becomes

$$

````r
loop_time = 10000 # samples repeatedly take
Distribution_plot <- function(distribution = 1, n = 10){
  sample <- c() # initialize the sample vector
  if (distribution == 1){
    title = paste("Summation Distribution: n =", n) # set title
    m = 18.9*n # theoretical mean
    sd = sqrt(119.07*n) # theoretical standard deviation
    for (i in 1:loop_time){
      sample <- c(sample, sum(runif(n, 0, 37.8))) # simulated samples(sum)
    }
  }else if(distribution == 2){
    title = paste("Mean Distribution: n =", n) # set title
    m = 18.9 # theoretical mean
    sd = sqrt(119.07/n) # theoretical standard deviation
    for (i in 1:loop_time){
      sample <- c(sample, mean(runif(n, 0, 37.8))) # 10000 simulated samples(mean)
    }
  }else{
    print("only 1 and 2 for distribution parameter")
  }
  
  h <- hist(sample, breaks=100, plot=F) # hitogram infos
  plot(h, freq=FALSE, main=title, xlab="Samples") # plot histogram
  x = seq(m-4*sd, m+4*sd, by = 0.01) # set x border for theoretical values
  lines(x, dnorm(x, m, sd), lwd=2.5, col="red") # plot theoretical curve
  legend("topright", c("Simulated Histogram", "Theoretical Values"), lty=c(1,1), col=c("black", "red"), pch=c("",""), lwd=c(1, 2.5))
}
````

I use boxcox to detect the distribution between each column and joyjoy. 

````r
## boxcox model
getLambda <- function(original_df){
  # input: dataframe
  # output: boxcox lambdas for each columns
  
  lambdas = c()
  for(i in names(original_df)[names(original_df) != "joyjoy"]){
    b <- boxcox(original_df$'joyjoy'~as.numeric(unlist(original_df[i]))) # boxcox transformation
    lambda <- b$x[which.max(b$y)] # get lambda
    lambdas <- c(lambdas, lambda) # append
  }
  return(lambdas)
}

getBoxcox_df <- function(original_df, lambdas){
  # input: dataframe, lambdas
  # output: transformed dataframe
  
  out_df <- data.frame() # initialize output dataframe
  count <- 1 # column index
  for(i in names(original_df)[names(original_df) != "joyjoy"]){
    new_x_exact <- ((original_df %>% select(i)) ^ lambdas[count] - 1) / lambdas[count] # transformed series
    count <- count + 1
    original_df[i] <- new_x_exact # replace original values with transformed values
  }
  original_df["joyjoy"] <- log(original_df["joyjoy"]) # log transformation on joyjoy

  return(original_df)
}

lambdas <- getLambda(train) # get lambdas
train_boxcox <- getBoxcox_df(original_df=train, lambdas) # boxcox transform train
val_boxcox <- getBoxcox_df(original_df=val, lambdas) # boxcox transform validation
````

````r
## test data
get_joyjoy <- function(test){
  ## boxcox transformation
  test <- test %>% mutate(joyjoy=0)
  test_boxcox <- getBoxcox_df(original_df=test, lambdas)
  test_predict <- predict(model_rf, test_boxcox, type = "class") # predict test cluster
  test_boxcox <- test_boxcox %>% mutate(cluster = test_predict)
  ## add columns (square data)
  test_sqr <- test^2 %>% select(-joyjoy)
  names(test_sqr) <- paste0(names(test_sqr), "_sqr")
  tmp <- test_boxcox %>% cbind(test_sqr)
  
  ## get predictions for different clusters
  # if cluster 0 is empty, then error
  tryCatch(
    {
      x_test0 <- tmp[tmp$cluster == 0, c(1:30, 33:62)] # predictors only
      pred_xgb0 <- predict(xgb0, as.matrix(x_test0)) # predict
      x_test0$joyjoy = pred_xgb0 # add predicion to a new columns
    },
    error = function(e, flag){
    }
  )

  # if cluster 1 is empty, then error
  tryCatch(
    {
      x_test1 <- tmp[tmp$cluster == 1, c(1:30, 33:62)]
      pred_xgb1 <- predict(xgb1, as.matrix(x_test1))
      x_test1$joyjoy = pred_xgb1
    },
    error = function(e, flag){
    }
  )

  print(dim(x_test0))
  print(dim(x_test1))
  ## rbind predictions of two clusters together
  if(dim(x_test1)[1] == 0){
    df <- x_test0 %>% select(joyjoy)
  }else if(dim(x_test0)[1] == 0){
    df <- x_test1 %>% select(joyjoy)
  }else{
    df <- rbind(x_test0%>%select(joyjoy), x_test1%>%select(joyjoy))
  }
  
  df$index <- as.numeric(row.names(df))
  pred_joyjoy = df[order(df$index), ]$joyjoy # sort by index
  return(exp(pred_joyjoy)) # return predictions from log scale back to original scale
}


````
