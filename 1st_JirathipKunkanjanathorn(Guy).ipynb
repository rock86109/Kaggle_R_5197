(a) The maximum likelihood estimator for the parameter $\theta$, i.e., $\hat{\theta}_{\text{MLE}}$ can be obtained as follows:

The likelihood is
$$
f(\textbf{x} \mid \theta)=f(x_1,x_2,...,x_n \mid \theta)=\prod_{i=1}^n f(x_i \mid \theta) = \left(\frac{1}{\theta}\right)^n.
$$
The negative log likelihood is 
$$
L(\textbf{x} \mid \theta)=-\log(f(\textbf{x} \mid \theta)) = -\log\left(\left(\frac{1}{\theta}\right)^n\right) = n\log(\theta).
$$
When we attempt to find the stationary point of the negative log likelihood we find that the derivative with respect to $\theta$ is
$$
\frac{dL(\textbf{x} \mid \theta)}{d\theta}=\frac{dn\log(\theta)}{d\theta}=\frac{n}{\theta}.
$$
We can't set this to zero to find the stationary point and solve for the estimate of $\theta$ because there is no stationary point. Instead we have to recognise that all of our realisations $x_1,x_2,...,x_n$ will be less than or equal to $\theta$ according to the PDF definition. This is equivalent to saying $\theta \geq x_1,x_2,...,x_n$. This means the negative log likelihood is actually defined with a specific domain as follows:
The negative log likelihood is 
$$
L(\textbf{x} \mid \theta)= n\log(\theta), \quad \theta \geq x_1,x_2,...,x_n.
$$
This is an increasing function with positive slope because $n$ and $\theta$ are positive and so the derivative of the function is also positive. We can also note the domain of this function starts at $\theta$ equal to the maximum value of $x_1,x_2,...,x_n$ since $\theta \geq x_1,x_2,...,x_n$. Therefore, since this is an increasing function starting from $\theta$ equal to the maximum value of $x_1,x_2,...,x_n$, the minimum of this function occurs when $\theta$ equals the maximum value of $x_1,x_2,...,x_n$. Therefore we let the MLE estimate be $\hat{\theta}=\text{argmax}_{x_i}\{x_1,x_2,...,x_n\}$. (Draw a diagram of the negative log likelihood if you don't get it.)

### ANSWER
(a) The likelihood function is the product of probability density functions

\begin{align*}
p({\bf x} | \theta) &= p(x_1,x_2,...,x_n \mid \theta)\\
&= \prod_{i=1}^n p(x_i | \theta)
\end{align*}
where
$p(x_i | \theta) = f(x_i | \theta) = \frac{\theta}{x_i^{1+\theta}}, \quad 1 \leq x_i<\infty ; \quad \theta>0$

The likelihood function is equivalent to:
\begin{align*}
p({\bf x} | \theta) &= \prod_{i=1}^n \frac{\theta}{x_i^{1+\theta}}\\
&= \theta ^ n x_i ^ {\sum_{i=1}^n -1-\theta}
\end{align*}

Therefore, maximum likelihood estimator ($\hat{\theta}_{\text{MLE}}$) can be found by solving for:

\begin{align*}
\hat{\theta}_{\text{MLE}} &= \text{argmax}_{\theta}\left\{p({\bf x} | \theta)\right\}\\
&= \text{argmax}_{\theta}\left\{ \prod_{i=1}^n p(x_i | \theta)\right\}\\
\end{align*}

Negative logarithm can be applied to both side to transform product of probability density function (likelihood function) into summation before breaking down the formula in previous part using the facts that:

(i) $\log (ab) = \log a + \log b$ and

(ii) $\log a^n = n \log a$

\begin{align*}
L(\textbf{x} \mid \theta) &= -\sum_{i=1}^n ln(p(x_i | \theta))\\
& = -\sum_{i=1}^n ln(\theta x_i^{-(1+\theta)})\\
& = -\sum_{i=1}^n ln(\theta) + ln(x_i^{-(1+\theta)})\\
&= -nln(\theta) + \sum_{i=1}^n (1+\theta)ln(x_i)\\
\end{align*}

This turns the maximum likelihood estimator ($\hat{\theta}_{\text{MLE}}$) to be:
\begin{align*}
\hat{\theta}_{\text{MLE}} &= \text{argmin}_{\theta}\left\{-\sum_{i=1}^n ln(p(x_i | \theta))\right\}
\end{align*}

where ${\text{argmin}_x \left\{ p(x) \right\}}$ means finding the value $x$ that minimises $p(x)$. Thus, minimises value of $p(x)$ can be obtained when the differentiation of negative log likelihood function with respect to parameter of interest ($\theta$).

\begin{align*}
\frac{\partial}{\partial \theta}L(\textbf{x} \mid \theta) &= \frac{\partial}{\partial \theta} \left(-nln(\theta) + \sum_{i=1}^n (1+\theta)ln(x_i)\right)\\
&= -\frac{n}{\theta} + \sum_{i=1}^n ln(x_i)\\
\end{align*}

Then set derivative of negative log likelihood function to zero and solve for $\theta$:

\begin{align*}
\hat{\theta}_{\text{MLE}} &= \frac{n}{\sum_{i=1}^n ln(x_i)}
\end{align*}

$$

\begin{align*}
\mathbb{E}[\hat{\theta}_{sample}] &= \mathbb{E}\left[\frac{1}{n}\sum_{i=1}^n {x_i} \right]\\
&= \frac{1}{n}\sum_{i=1}^n \mathbb{E}[{x_i}]\\
\end{align*}


However, the expectation of $x_i$ is:
$$

$$
where
$p(x_i | \theta) = f(x_i | \theta) = \frac{\theta}{x_i^{1+\theta}}, \quad 1 \leq x_i<\infty ; \quad \theta>0$

Thus 
\begin{align*}

## Question 1: Theoretical Set-up for the CLT (No Coding or Simulation here!) (2 Marks)

### ANSWER
From Central Limit Theorem (CLT), The summation of many random variables that are independent and idenfically distributed with finite mean and variance of any distribution converges to normal distribution as n approach infinity $n \rightarrow \infty$:
$$
\sum_{i}^{n}X_{i} \stackrel{d}{\rightarrow} \mathcal{N}\left(n\mu, n\sigma^{2}\right)
$$
and the mean of distribution converges to:
$$
\overline{X} \stackrel{d}{\rightarrow} \mathcal{N}\left(\mu, \sigma^{2} / n\right)
$$

The plug-in distribution A has probability density function of:
$$
f(x \mid \hat{\theta}=37.8)=\frac{1}{37.8}, \quad 0 \leq x \leq 37.8.
$$
Thus, the expected value is:
\begin{align*}
\mathbb{E}[{X}] &= \int_{x\in X} \frac{x}{37.8} \thinspace dx\\
&= \int_{0}^{37.8} \frac{x}{37.8} \thinspace dx\\
&= \frac{x^2}{2\times37.8} \Big|_{0}^{37.8}\\
&= \frac{37.8}{2}\\
&= 18.9
\end{align*}
and the variance is:
$$
\mathbb{V}[{X}] = \mathbb{E}[{X^2}] - \mathbb{E}[{X}]^2
$$
where expected value of $X^2$ is:
\begin{align*}
\mathbb{E}[{X^2}] &= \int_{0}^{37.8} \frac{x^2}{37.8} \thinspace dx\\
&= \frac{x^3}{3\times37.8}\Big|_{0}^{37.8}\\
&= 476.28
\end{align*}

$$
\mathbb{V}[{X}] = 476.28 - 18.9^2 = 119.07
$$


Since the distribution is actually uniform distribution $\operatorname{U}(0, 37.8)$, this is also equal to the closed form of expected value and variance:
$$
\mathbb{E}[{X}] = \frac{a+b}{2}
= \frac{37.8}{2} = 18.9
$$
$$
\mathbb{V}[{X}] = \frac{(b-a)^2}{12} = \frac{37.8^2}{12} = 119.07
$$

Therefore, the distribution for plug-in distribution A in form of **Summation Distribution** and **Mean Distribution** for $\;\operatorname{n}_{\operatorname{Small}} = 10\;$, $\;\operatorname{n}_{\operatorname{Medium}} = 30\;$, $\;\operatorname{n}_{\operatorname{Big}} = 100\;$ are as follow:

    - $\;\operatorname{n}_{\operatorname{Medium}} = 30\;$: $\sim \mathcal{N}\left(30\times18.9, 30\times119.07\right)$

````r
# ANSWER BLOCK
# Create function in generate repeated uniform random variable sample
sample.uniform <- function(nsampling, n, min, max) {
    # Initiate empty list and x grid for generating and ploting normal curve
    result <- vector("list")
    result$n <- n
    result$x.grid.mean <- seq(min, max, length = nsampling)
    result$x.grid.sum <- seq(min, max * n, length = nsampling)
    
    # Store theoretical mean and variance for each cases
    result$t.mean.mean <- (max + min) / 2
    result$t.mean.sum <- result$t.mean * n
    
    result$t.var.mean <- ((max - min) ^ 2) / (12 * n)
    result$t.var.sum <- result$t.var * n^2
    
    # Theoretical normal curve
    result$normal.curve.mean <- dnorm(result$x.grid.mean, 
                                      mean=result$t.mean.mean, 
                                      sd= sqrt(result$t.var.mean))
    result$normal.curve.sum <- dnorm(result$x.grid.sum, 
                                     mean=result$t.mean.sum, 
                                     sd= sqrt(result$t.var.sum))
    
    # Repeat sampling, calculate statistic, and store in list
    for (sample in 1:nsampling) {
        
        # a continuous uniform random variables
        x <- runif(n = n, min = min, max = max)
        
        # Calculate sum and mean of the random variables
        result$sum[sample] <- sum(x)
        result$mean <- result$sum / n
        }
    return (result)
}
````

````r
# Generate repeated sample of n equal to 10, 30, and 100 for uniform random variables b = 37.8 and a = 0
sampling.result <- lapply(c(10, 30, 100), sample.uniform, nsampling = 10000, min = 0, max = 37.8)
````

````r
# ANSWER BLOCK
least_square <- function(df, p){ # df is the dataframe; p is the predictor name
    ## implement the least square estimator here
    
    # Calculate sample statistics for solution for Beta 0 hat and Beta 1 hat
    x.bar <- mean(df[[p]])
    x.square.bar <- mean(df[[p]]^2)
    y.bar <- mean(df[['joyjoy']])
    x.y.bar <- mean(df[[p]] * df[['joyjoy']])
    
    # Calculate Beta 0 hat and Beta 1 hat
    beta_0_hat <- (y.bar * x.square.bar - x.y.bar * x.bar) / (x.square.bar - x.bar ^ 2)
    beta_1_hat <- (x.y.bar - x.bar * y.bar) / (x.square.bar - x.bar ^ 2)
    
    print(paste0('E[joyjoy]=',beta_0_hat,'+',beta_1_hat,'*',p)) 
}

least_square(train, 'Negativesocialskillstotal')
````

Moreover, hypothesis testing will be applied to test with null hypothesis that they are from the same distribution with assumption that they are Gaussian distribution with unknown and unequal variance. 

$$

$$
$$
H_A: \mu_x \neq \mu_y

$$
\hat\mu_x -\hat\mu_y \sim \mathcal{N} \left(0, \frac{\sigma_x^2}{n_x} + \frac{\sigma_y^2}{n_y}\right)
$$

And test statistic z-score of the difference in the means is:

\begin{align*}
z_{\hat\mu_x -\hat\mu_y} &= \frac{\bar x - \bar y}{\sqrt{\frac{s_x^2}{n_x} + \frac{s_y^2}{n_y}}}\\
&= \frac{35.53 - 61.95}{\sqrt{\frac{11.85^2}{40000} + \frac{20.69^2}{10000}}}\\
&= -122.76

## Feature Engineering using Tree based model

From exploratory data analysis in the previous part, hidden group has been identified. Therefore, classification models will be applied to group the data and test that grouping can be predicted using other independent variables exclude `joyjoy`. Decision tree is choosen for this classification as the dataset seems to be non-linear and the tree requires less assumption about the data. The tree is splitting the data based on the impurity or the variance of the nodes starting with the nodes with lowest impurity. In this model, information entropy is used where the entropy is

$$
H(X) = \mathbb{E}\left[ ( log_2 \frac{1}{p_i(X)})\right]
$$

where $p_i$ is a probability of belonging to class $i$. The tree is splitting until cost of adding another variable is above the value of complexity parameter (cp) which is the minimum improvement for each node added. The complexity parameter was set to 0.001 in this model to grow a tree. `printcp` is indicated that this hyper parameter resulting in 29 splits in this model with relative error ($1-R^2$) of 0.005. Confusion matrix is created to confirm the result which can be seen that only 55 out of 10,000 observations is predicted incorrectly. F1 score is also calculated below.

````r
# Remove joyjoy when predicting hidden group
train.ex.joyjoy <- subset(train, select=-joyjoy)

# Using decision tree to classify hidden group
classify.group <- rpart(group ~., data = train.ex.joyjoy, 
                        parms = list(split = "information"), 
                        control = rpart.control(cp = 0.001))
# Print model result
printcp(classify.group)
````

````r
# Create confusion matrix to verify result
confusionMatrix(predict(classify.group, train, type = 'class'), train$group,
                mode = 'everything',
                positive='1')
````

## Adding Polynomial Terms and Hidden Grouping

Polynomial models up to degree five with 31 original predictors with group variable is compared using RMSE and cross validation errors. The function for the regression model is as follows:
$$
\vec{y} = X\vec{\beta} +\beta_g\vec{g} + \vec{\epsilon}
$$
where $X$ is design matrix
$$
X = 
\begin{bmatrix}
1 & x_1 & x_1^2 & \dots & x_1^m \\
1 & x_2 & x_2^2 & \dots & x_2^m \\
\dots  & \dots & \dots  & \dots & \dots  \\
1 & x_n & x_n^2 & \dots & x_n^m 
\end{bmatrix}
$$
with $\vec{y}$, $\vec{\beta}$,$\vec{g}$ , and $\vec{\epsilon}$ are vector of target observations, coefficients, grouping (categorical) obsevations, and random errors respectively
$$
\vec{y}
=
\begin{bmatrix}
y_1 \\ y_2 \\ \dots \\ y_n 
\end{bmatrix}
,\;
\vec{\beta} = 
\begin{bmatrix}
\beta_1 \\ \beta_2 \\ \dots \\ \beta_n
\end{bmatrix}
,\;
\vec{g} = 
\begin{bmatrix}
g_1 \\ g_2 \\ \dots \\ g_n
\end{bmatrix}
,\;
\vec{\epsilon} = 
\begin{bmatrix}
\epsilon_1 \\ \epsilon_2 \\ \dots \\ \epsilon_n
\end{bmatrix}
$$
and $\beta_g$ is coeficients for categorical variables. It should be noted that categorical variable does not include in polynomial terms.

n = number of predictors

m = degree of polynomials

````r
# Store predictors (features) exclude group in a variable
features <- colnames(train.ex.joyjoy)[1:ncol(train.ex.joyjoy) - 1]

# Function to create model formula for each degree
poly.formula <- function(degree.poly, target) {
    # Create empty list to store formula
    formula.poly <- vector('list', degree.poly)
    
    # Loop to generate formula for linear regression for each polynomial degree
    for (i in 1:degree.poly){
            formula.poly[[i]] <- as.formula(
                paste(paste0(target," ~ "),
                      paste('poly(',features, ', ', i,', raw = TRUE)',collapse = '+'),
                      '+group', 
                      sep = ''))
        }
    return (formula.poly)
}

# Function to calculate metric for each degree
poly.rmse <- function(degree.poly, target, metric) {
    
    # Create polynomial formulas
    formula.poly <- poly.formula(degree.poly, target)
    
    # Create empty lists to store model rMSE
    rmse.list <- vector('list', degree.poly)
    rmse.test.list <- vector('list', degree.poly)
    
    for (i in 1:degree.poly){
        # Train the model
        fit <- lm(formula.poly[[i]], data = new.train)
        
        # Calculate metric
        rmse.list[[i]] <- metric(new.train, fit)
        rmse.test.list[[i]] <- metric(new.test, fit)
        
        }

    # Format result to dataframe
    model.metric <- do.call(rbind, Map(data.frame,
                                       rmse = rmse.list,
                                       rmse.test = rmse.test.list))
    return (model.metric)
    }
````

## Transform dependent variable using Box-Cox transformation

From the result of polynomial regressions, it can be seen that metrics such as R-square and RMSE improved dramatically from feature engineering which includes adding hidden groups and polynomial terms in previous part. However, problems was detected from the diagnostic plots above. It has revealed that there is heterogeneous in the residuals and the dependent variable is not gaussian. Box-Cox transformation is choosen to alleviate these problems to make the data more normal distribution-like. All lambda ($\lambda$) value from range of -5 to 5 is considered for the optimal value that best approximate normal distribution in equation as follows:

\begin{equation*}
    y(\lambda) =\left\{
                \begin{array}{ll}
                   \frac{y^{\lambda}-1}{\lambda} \;\text{ if }\;\lambda \neq 0\\
                   log(y) \;\text{ if  }\;\lambda = 0
                \end{array}
              \right.
\end{equation*}

It should be noted that the lambda to search is set in the range of -1 to 1 by 0.02 to optimise performance of the model.

````r
# Call boxcox function from MASS library to search for optimal lambda that transform joyjoy to normal distribution
bc <- boxcox(poly.formula(3, 'joyjoy')[[3]], data = train,
             lambda = seq(-1, 1, by = 0.02))
# Set plot title for Box Cox plot
title(main = "Box Cox transformation optimal lambda")

# Assign lambda value from the boxcox model to variable
lambda <- bc$x[which.max(bc$y)]

lambda
````

````r
# Create function to transform result from boxcox transform prediction to the original data scale
power.transform.prediction <- function(model, df){
    return (((predict(model, df) * lambda) + 1)^(1/lambda))
}

# Create function to calculate RMSE of transform result from boxcox transfrom prediction
rmse.power.transform <- function(df,model){
    ## implement rMSE squared here
    
    y <- df$joyjoy
    y.hat <- power.transform.prediction(model, df)
    
    return (sqrt(mean((y - y.hat)^2)))
}
````

````r
# Test the RMSE for each degree after dependent variable is transformed
degree.poly <- 8
fit.formula.poly.bc <- poly.formula(degree.poly, '((joyjoy^lambda-1)/lambda)')

poly.rmse(degree.poly, '((joyjoy^lambda-1)/lambda)', rmse.power.transform)
````

Since the koalas are in the first group $\frac{39,950}{40,000}$ of the times from random forest prediction. Regarding central limit theorem, approximation of $\hat \theta$ is 0.99875. Each observation is benoulli random variable. Therefore, hypothesis testing can be used to test that the probability of koalas in first group is 0.8 which is the probability of the first group in the training data.


$$
H_0: \theta = 0.8
$$
$$
H_A: \theta \neq 0.8
$$

\begin{align*}
z_{\hat\theta} &= \frac{\hat \theta - \theta}{\theta (1 - \theta) / n}\\
&= \frac{0.99875 - 0.8}{0.8(1-0.8)/ 50000}\\
&= 111.1
\end{align*}

The result is strongly rejected the null that the probability of koalas to be in the first group is 0.8 at significant level 0.05.

Moreover, it can also be tested that the result of the test set actually has the same distribution as the first group in training set that has mean ($\bar x$) and standard deviation ($s_x$) of 35.53311 and 11.84969 respectively. The sample statistics of the prediction result $\bar y$ and $s_y$ are 35.50984 and 11.7903 respectively. Assuming that variances of the training and testing set is the same.

$$
H_0: \mu_x = \mu_y
$$
$$
H_A: \mu_x \neq \mu_y
$$

the pool estimate of $\sigma^2$ is:

$$
s_p^2 = \frac{(n_x - 1)s_x^2 + (n_y - 1)s_y^2}{n_x + n_y - 2}
$$

With $n_x$ and $n_y$ = 40,000
$$
s_p^2 = 139.713
$$


By central limit theorem, test statistic t-score for difference between two mean is:

\begin{align*}
t_{\hat\mu_x -\hat\mu_y} &= \frac{\bar x - \bar y}{\sqrt{s_p^2 (1 / n_x + 1/ n_y)}}\\
&= \frac{35.53 - 35.51}{\sqrt{139.713 (1/40000 + 1/40000)}}\\
&= 0.24
\end{align*}


The t-score of 0.24 with degree of freedom 79,998 has p-value of 0.81. Therefore, there is not enough evidence to reject the null hypothesis that they are the same with significant level 0.05. Thus, given the information from hypothesis testing above that the prediction from random forest predicting 50 out of 40,000 observations are from the second group (1). This is likely to be false negative and all of the observations are from first group (0). With assumption that all of them are the first group, grouping will be replaced with the first group (0) for all observations and predict again.

````r
# ANSWER BLOCK
# Get only importance polynomial predictors from the model
all.imp.coefficients.poly <- all.coefficients.poly %>%
    filter(`Pr(>|t|)` < 0.01) %>%
    rownames_to_column() %>%
    filter(rowname != '(Intercept)') %>%
    mutate(imp.variable = str_replace(str_replace(str_extract(rowname, "poly.*?,"),",",""),"poly", "")) %>%
    mutate(imp.variable = substr(imp.variable, 2, length(imp.variable))) %>%
    mutate(degree = substr(rowname, nchar(rowname), nchar(rowname))) %>%
    filter(imp.variable %in% imp.variable) %>%
    dplyr::select(imp.variable, degree, Estimate) %>% 
    drop_na() %>%
    pivot_wider(names_from = degree, values_from = Estimate)
    

get.var.func <- function (var.name) {
    var <- all.imp.coefficients.poly[all.imp.coefficients.poly$imp.variable == var.name,]
    coef <- c(var$`1`, var$`2`, var$`3`, var$`4`)
    var.func <- function (x) coef[1] * x + coef[2] * x^2 + coef[3] * x^3 + coef[4] * x^4
    return (var.func)
}

var.func.list <- vector("list")
for (i in 1:length(imp.variable$imp.variable)){
    var.func.list[[imp.variable$imp.variable[i]]] <- get.var.func(imp.variable$imp.variable[i])
}

plot <- ggplot() + xlim(0,100)
plot + lapply(var.func.list, function(x) {
    stat_function(fun = x)
}) +
stat_function(fun = get.var.func('Socialadaptation'), aes(color = 'Socialadaptation')) +
stat_function(fun = get.var.func('Cooperativebehavior'), aes(color = 'Cooperativebehavior')) +
labs(title = "Polynomial function of each importance coefficients",
     subtitle = 'Effect of each coeficient estimate of predictors on joyjoy transformed\nfor each x unit from 0 to 100',
     x = 'x', 
     y = 'marginal change for each unit of x for transform joyjoy',
     color = "Predictors") +
theme(legend.position = "bottom")
````
